```python
# Importe de librerias
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import contextily as cx
from pathlib import Path
import pandas as pd
from shapely.geometry import Point
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
from matplotlib.patches import Patch
import numpy as np
import shapely
import rasterio
import rasterio.plot
from rasterio.mask import mask
from shapely.geometry import mapping
from rasterio.warp import calculate_default_transform, reproject, Resampling
```

```python
RAS=r"/notebooks/Proyecto_Clase/results/"
ruta_lectura_DEM = RAS+r"DEM_Cortado_Curiti_Buff.tif"

Dem_cortado = rasterio.open(ruta_lectura_DEM)
Dem_data = Dem_cortado.read(1)

valor_nodata_archivo = Dem_cortado.nodata

print("Valor Nodata:", Dem_cortado.nodata)

Dem_data = np.ma.masked_equal(Dem_data, valor_nodata_archivo)
# Plotear el recorte

plt.figure(figsize=(8, 6))
plt.imshow(Dem_data, cmap='terrain')            
plt.colorbar(label="Elevacion [metros]")       
plt.title("Visualizacion DEM Cortado")                        
plt.axis("off")                                 
plt.show()
```
