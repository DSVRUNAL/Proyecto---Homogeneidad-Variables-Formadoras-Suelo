## Ejecución de funcion k-means paralelizada por cada símbolo de la geoforma
1. Carga las rutas del shape y de los raster recortados
2. Ejecuta el k-means iterando por cada símbolo de geoforma
3. Se obtiene como resultado un dataframe con los valores k y Silhouette Score óptimos para cada símbolo

```python
import sys
sys.path.append("/notebooks/Proyecto_Clase/src")
import symbol_processing
importlib.reload(symbol_processing)
import geopandas as gpd
import matplotlib.pyplot as plt

# Define las rutas y los nombres de los archivos
ruta_shp = "/notebooks/Insumos_Proyecto_Clase/"
ruta_ras = "/notebooks/Insumos_Proyecto_Clase/Raster_Cortados/"
nombres_shp = {
    "geom": "Geomorfologia_clip_igac.shp"
}

# Llama a la función para procesar los símbolos y obtener el resultado final
tabla_final = symbol_processing.procesar_simbolos(ruta_shp, ruta_ras, nombres_shp)
```

## Calificación de Geoformas y Presentación final de su clasificación
1. Se califica el dataframe resultante con los umbrales establecidos: K=2 & k<=5 y SS>0.87 para geoformas homogéneas
2. Se hace un merge contra el geodataframe de las geoformas
3. Se grafica la clasificación final

```python
def calificar(row):
    if 2 <= row['Promedio_k'] <= 5 and row['Promedio_SS'] > 0.87:
        return 'Homogenea'
    else:
        return 'Heterogenea'

# Calificar el dataframe adicionando una columna de calificacion
tabla_final["Calificacion"] = tabla_final.apply(calificar, axis=1)

# Mostrar el resultado completo
print(tabla_final)
```

```python

geom_shp = gpd.read_file('/notebooks/Insumos_Proyecto_Clase/Geomorfologia_clip_igac.shp')
geoformas_clas = geom_shp.merge(
    tabla_final,
    how='left',                 # unir evas a municipios
    left_on='Simbolo',          # columna clave en municipios
    right_on='Simbolo',         # columna clave en evas
    indicator=True,             # crea indicador de unión en columna nueva
    validate='many_to_many',     # relación uno-muchos para que se unan todos los registros de cultivos por municipio
    suffixes=('', '_Tfinal')      # añade sufijos cuando existan nombres de columna iguales
)

print(geoformas_clas)
```

```python

# Plotear el mapa de calificación por geoforma
fig, ax = plt.subplots(1, 1, figsize=(8, 8))

# Generar el mapa utilizando el atributo 'Calificación' para colorear las geoformas
geoformas_clas.plot(column='Calificacion', 
                    k=2,            # Dos colores, uno para homogéneas y otro para heterogéneas
                   cmap='coolwarm', # Puedes cambiar el cmap a tu preferencia
                   legend=True, 
                   ax=ax, 
                   edgecolor='grey', 
                   linewidth=0.2)

# Título del gráfico
ax.set_title('Mapa de Calificación de Geoformas', fontsize=16)
ax.set_axis_off()  # Ocultar los ejes
plt.show()
```
